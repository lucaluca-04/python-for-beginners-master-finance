### Looping Over A File Object In Python: 

---


When you want to read â€“ or return â€“ all the lines from a file in a more memory efficient, and fast manner, you can use the loop over method. The advantage to using this method is that the related code is both simple and easy to read. 



```python
file = open("demo.txt", "r") 
for line in file: 
  print(line)
```

## File Write Method In Python : 

---
One thing youâ€™ll notice about the file write method is that it only requires a single parameter, which is the string you want to be written.

This method is used to add information or content to an existing file. To start a new line after you write data to the file, you can add an EOL character.

```python
file = open("demo.txt", "w")
 
file.write("This is a test") 
file.write("To add more lines.")
 
file.close()
```

### Closing A File In Python: 

---


When youâ€™re done working, you can use the fh.close() command to end things. What this does is close the file completely, terminating resources in use, in turn freeing them up for the system to deploy elsewhere.

Itâ€™s important to understand that when you use the fh.close() method, any further attempts to use the file object will fail.

### 1. CSV æ–‡ä»¶æ“ä½œ - è‚¡ç¥¨æ•°æ®

CSVï¼ˆé€—å·åˆ†éš”å€¼ï¼‰æ˜¯é‡‘èæ•°æ®æœ€å¸¸ç”¨çš„æ ¼å¼ï¼Œç‰¹åˆ«æ˜¯è‚¡ç¥¨ä»·æ ¼æ•°æ®ã€‚

```python
# åˆ›å»ºè‚¡ç¥¨æ•°æ®çš„CSVæ–‡ä»¶
import csv

# å‡†å¤‡è‚¡ç¥¨æ•°æ®
stock_data = [
    ['Date', 'Open', 'High', 'Low', 'Close', 'Volume'],
    ['2024-01-01', '100.5', '102.0', '99.8', '101.2', '1000000'],
    ['2024-01-02', '101.2', '103.5', '100.5', '102.8', '1200000'],
    ['2024-01-03', '102.8', '104.0', '101.5', '103.2', '1100000'],
    ['2024-01-04', '103.2', '105.0', '102.0', '104.5', '1300000'],
    ['2024-01-05', '104.5', '106.0', '103.5', '105.2', '1250000']
]

# å†™å…¥CSVæ–‡ä»¶
with open('stock_data.csv', 'w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerows(stock_data)

print("è‚¡ç¥¨æ•°æ®å·²ä¿å­˜åˆ° stock_data.csv")
```

### 2. ä½¿ç”¨ Pandas å¤„ç†é‡‘èæ•°æ®æ–‡ä»¶

Pandas æ˜¯å¤„ç†é‡‘èæ•°æ®æœ€å¼ºå¤§çš„å·¥å…·ï¼Œå¯ä»¥è½»æ¾è¯»å–å’Œä¿å­˜å„ç§æ ¼å¼çš„æ•°æ®ã€‚

---

```python
# ä½¿ç”¨ pandas è¯»å– CSV æ–‡ä»¶
import pandas as pd
import numpy as np

# è¯»å–è‚¡ç¥¨æ•°æ®
df = pd.read_csv('stock_data.csv')

print("ä½¿ç”¨ pandas è¯»å–çš„æ•°æ®:")
print(df)
print(f"\næ•°æ®ç±»å‹:\n{df.dtypes}")
print(f"\næ•°æ®ç»Ÿè®¡:\n{df.describe()}")
```

```python
# å¤„ç†æ—¥æœŸåˆ—å¹¶è®¾ç½®ä¸ºç´¢å¼•
df['Date'] = pd.to_datetime(df['Date'])
df = df.set_index('Date')

print("è®¾ç½®æ—¥æœŸä¸ºç´¢å¼•å:")
print(df)

# è®¡ç®—æ”¶ç›Šç‡
df['Return'] = df['Close'].pct_change()
df['MA5'] = df['Close'].rolling(window=3).mean()

print("\næ·»åŠ æŠ€æœ¯æŒ‡æ ‡å:")
print(df[['Close', 'Return', 'MA5']])
```

```python
# ä¿å­˜å¤„ç†åçš„æ•°æ®åˆ° CSV
df.to_csv('stock_data_processed.csv')
print("å¤„ç†åçš„æ•°æ®å·²ä¿å­˜åˆ° stock_data_processed.csv")

# ä¿å­˜ä¸º Excel æ ¼å¼ï¼ˆéœ€è¦å®‰è£… openpyxl: pip install openpyxlï¼‰
try:
    df.to_excel('stock_data.xlsx')
    print("æ•°æ®å·²ä¿å­˜ä¸º Excel æ ¼å¼: stock_data.xlsx")
except ImportError:
    print("æç¤º: éœ€è¦å®‰è£… openpyxl æ‰èƒ½ä¿å­˜ Excel æ–‡ä»¶: pip install openpyxl")
```

### 3. JSON æ–‡ä»¶æ“ä½œ - é‡‘èæ•°æ®

JSON æ ¼å¼å¸¸ç”¨äº API è¿”å›çš„é‡‘èæ•°æ®ã€‚

```python
# åˆ›å»ºé‡‘èæ•°æ®çš„ JSON æ–‡ä»¶
import json

# æ¨¡æ‹Ÿä» API è·å–çš„é‡‘èæ•°æ®
financial_data = {
    "symbol": "AAPL",
    "company": "Apple Inc.",
    "prices": [
        {"date": "2024-01-01", "open": 100.5, "high": 102.0, "low": 99.8, "close": 101.2, "volume": 1000000},
        {"date": "2024-01-02", "open": 101.2, "high": 103.5, "low": 100.5, "close": 102.8, "volume": 1200000},
        {"date": "2024-01-03", "open": 102.8, "high": 104.0, "low": 101.5, "close": 103.2, "volume": 1100000}
    ],
    "metadata": {
        "currency": "USD",
        "exchange": "NASDAQ",
        "timezone": "America/New_York"
    }
}

# å†™å…¥ JSON æ–‡ä»¶
with open('financial_data.json', 'w', encoding='utf-8') as file:
    json.dump(financial_data, file, indent=2, ensure_ascii=False)

print("é‡‘èæ•°æ®å·²ä¿å­˜åˆ° financial_data.json")
```

```python
# è¯»å– JSON æ–‡ä»¶
with open('financial_data.json', 'r', encoding='utf-8') as file:
    data = json.load(file)

print("è¯»å–çš„é‡‘èæ•°æ®:")
print(f"è‚¡ç¥¨ä»£ç : {data['symbol']}")
print(f"å…¬å¸åç§°: {data['company']}")
print(f"\nä»·æ ¼æ•°æ®:")
for price in data['prices']:
    print(f"æ—¥æœŸ: {price['date']}, æ”¶ç›˜ä»·: {price['close']}")
```

### 4. ä½¿ç”¨ with è¯­å¥ï¼ˆæ¨èæ–¹å¼ï¼‰

ä½¿ç”¨ `with` è¯­å¥å¯ä»¥è‡ªåŠ¨å…³é—­æ–‡ä»¶ï¼Œé¿å…å¿˜è®°å…³é—­æ–‡ä»¶çš„é—®é¢˜ã€‚

```python
# âœ… æ¨èæ–¹å¼ï¼šä½¿ç”¨ with è¯­å¥
# æ–‡ä»¶ä¼šè‡ªåŠ¨å…³é—­ï¼Œä¸éœ€è¦æ‰‹åŠ¨è°ƒç”¨ close()

# å†™å…¥æ–‡ä»¶
with open('trading_log.txt', 'w', encoding='utf-8') as file:
    file.write("äº¤æ˜“æ—¥å¿—\n")
    file.write("=" * 30 + "\n")
    file.write("2024-01-01 09:30:00 - ä¹°å…¥ AAPL 100è‚¡ @ $101.2\n")
    file.write("2024-01-01 15:00:00 - å–å‡º AAPL 100è‚¡ @ $102.5\n")
    file.write("ç›ˆåˆ©: $130\n")

print("äº¤æ˜“æ—¥å¿—å·²ä¿å­˜")

# è¯»å–æ–‡ä»¶
with open('trading_log.txt', 'r', encoding='utf-8') as file:
    content = file.read()
    print("\näº¤æ˜“æ—¥å¿—å†…å®¹:")
    print(content)
```

### 5. è¿½åŠ æ¨¡å¼ - è®°å½•äº¤æ˜“æ•°æ®

ä½¿ç”¨è¿½åŠ æ¨¡å¼ï¼ˆ'a'ï¼‰å¯ä»¥åœ¨æ–‡ä»¶æœ«å°¾æ·»åŠ æ–°æ•°æ®ï¼Œé€‚åˆè®°å½•äº¤æ˜“æ—¥å¿—ã€‚

---

```python
# è¿½åŠ æ¨¡å¼å†™å…¥äº¤æ˜“è®°å½•
from datetime import datetime

def log_trade(symbol, action, quantity, price):
    """è®°å½•äº¤æ˜“åˆ°æ—¥å¿—æ–‡ä»¶"""
    timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    with open('trading_log.txt', 'a', encoding='utf-8') as file:
        file.write(f"{timestamp} - {action} {symbol} {quantity}è‚¡ @ ${price}\n")

# è®°å½•å‡ ç¬”äº¤æ˜“
log_trade("AAPL", "ä¹°å…¥", 100, 101.2)
log_trade("MSFT", "ä¹°å…¥", 50, 250.5)
log_trade("AAPL", "å–å‡º", 100, 102.5)

# è¯»å–æ‰€æœ‰äº¤æ˜“è®°å½•
with open('trading_log.txt', 'r', encoding='utf-8') as file:
    print("æ‰€æœ‰äº¤æ˜“è®°å½•:")
    print(file.read())
```

```python
# ç»¼åˆç¤ºä¾‹ï¼šå¤„ç†è‚¡ç¥¨æ•°æ®æ–‡ä»¶

# 1. åˆ›å»ºæ¨¡æ‹Ÿçš„è‚¡ç¥¨æ•°æ®
dates = pd.date_range('2024-01-01', periods=10, freq='D')
prices = 100 + np.cumsum(np.random.randn(10) * 2)

stock_df = pd.DataFrame({
    'Date': dates,
    'Open': prices + np.random.randn(10) * 0.5,
    'High': prices + abs(np.random.randn(10) * 1),
    'Low': prices - abs(np.random.randn(10) * 1),
    'Close': prices,
    'Volume': np.random.randint(1000000, 5000000, 10)
})

# 2. ä¿å­˜ä¸º CSV
stock_df.to_csv('daily_stock_data.csv', index=False)
print("è‚¡ç¥¨æ•°æ®å·²ä¿å­˜")

# 3. è¯»å–å¹¶å¤„ç†æ•°æ®
df = pd.read_csv('daily_stock_data.csv')
df['Date'] = pd.to_datetime(df['Date'])
df = df.set_index('Date')

# è®¡ç®—æŠ€æœ¯æŒ‡æ ‡
df['Return'] = df['Close'].pct_change()
df['MA5'] = df['Close'].rolling(window=5).mean()
df['Volatility'] = df['Return'].rolling(window=5).std()

# 4. ä¿å­˜å¤„ç†åçš„æ•°æ®
df.to_csv('processed_stock_data.csv')
print("\nå¤„ç†åçš„æ•°æ®:")
print(df[['Close', 'Return', 'MA5', 'Volatility']].head())
```

### 8. æœ€ä½³å®è·µå’Œæ³¨æ„äº‹é¡¹

**é‡è¦æç¤ºï¼š**

1. **å§‹ç»ˆä½¿ç”¨ `with` è¯­å¥**ï¼šè‡ªåŠ¨å…³é—­æ–‡ä»¶ï¼Œé¿å…èµ„æºæ³„æ¼
2. **æŒ‡å®šç¼–ç **ï¼šä½¿ç”¨ `encoding='utf-8'` å¤„ç†ä¸­æ–‡æ•°æ®
3. **ä½¿ç”¨ pandas å¤„ç†å¤§å‹æ•°æ®**ï¼šæ¯”çº¯ Python æ–‡ä»¶æ“ä½œæ›´é«˜æ•ˆ
4. **å¤‡ä»½é‡è¦æ•°æ®**ï¼šåœ¨è¦†ç›–æ–‡ä»¶å‰å…ˆå¤‡ä»½
5. **å¼‚å¸¸å¤„ç†**ï¼šä½¿ç”¨ try-except å¤„ç†æ–‡ä»¶ä¸å­˜åœ¨ç­‰é”™è¯¯

---

```python
# æœ€ä½³å®è·µç¤ºä¾‹ï¼šå¸¦å¼‚å¸¸å¤„ç†çš„æ–‡ä»¶æ“ä½œ

def safe_read_csv(filename):
    """å®‰å…¨åœ°è¯»å– CSV æ–‡ä»¶ï¼Œå¸¦å¼‚å¸¸å¤„ç†"""
    try:
        df = pd.read_csv(filename, encoding='utf-8')
        print(f"æˆåŠŸè¯»å–æ–‡ä»¶: {filename}")
        return df
    except FileNotFoundError:
        print(f"é”™è¯¯: æ–‡ä»¶ {filename} ä¸å­˜åœ¨")
        return None
    except Exception as e:
        print(f"è¯»å–æ–‡ä»¶æ—¶å‡ºé”™: {e}")
        return None

def safe_write_csv(df, filename):
    """å®‰å…¨åœ°å†™å…¥ CSV æ–‡ä»¶"""
    try:
        df.to_csv(filename, index=False, encoding='utf-8')
        print(f"æˆåŠŸä¿å­˜æ–‡ä»¶: {filename}")
        return True
    except Exception as e:
        print(f"ä¿å­˜æ–‡ä»¶æ—¶å‡ºé”™: {e}")
        return False

# ä½¿ç”¨ç¤ºä¾‹
df = safe_read_csv('stock_data.csv')
if df is not None:
    print(f"\næ•°æ®å½¢çŠ¶: {df.shape}")
    print(f"åˆ—å: {df.columns.tolist()}")
```
